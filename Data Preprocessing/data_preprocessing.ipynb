```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from collections import defaultdict
import time

weather_data = pd.read_csv("weather_data.csv")
```

```python
weather_data.head()
```

```python
wdir = []
prcp = []


for idx, row in weather_data.iterrows():
    if row['wspd'] == 0:
        wdir.append(0)
    else:
        wdir.append(row['wdir'])
        
    if math.isnan(row['prcp']):
        prcp.append(0)
    else:
        prcp.append(row['prcp'])
        
weather_data['wdir'] = wdir
weather_data['prcp'] = prcp
```

```python
weather_data.head()
weather_data.to_csv("weather_data.csv")
```

```python
lyft_data = pd.read_csv("lyft_data.csv")
lyft_data.head()
```

```python
date = []
time = []

for idx, row in lyft_data.iterrows():
    date.append(row['datetime'].split('_')[0])
    time.append(row['datetime'].split('_')[1])
    
lyft_data['date'] = date
lyft_data['time'] = time
        
lyft_data.head()
```

```python
parking_data = pd.read_csv("park_data_cleaned.csv")
parking_data.head()
muni_data = pd.read_csv("muni_data_cleaned.csv")
muni_data.head()
```

```python
parking_dict = defaultdict(float)  # zipcode -> acres
muni_dict = defaultdict(int)  # zipcode -> acres

for idx, row in parking_data.iterrows():
    parking_dict[row["Zipcode"]] = parking_dict[row["Zipcode"]] + row["Acres"]

for idx, row in muni_data.iterrows():
    muni_dict[row["zipcode"]] += 1
```

```python
weather_dict = defaultdict(list)  # zipcode -> acres

for idx, row in weather_data.iterrows():
    datetime_idx = str(row["date"]) + '_' + str(row["time"].split(':')[0])
    weather_dict[datetime_idx].append(row.get('temp', 0))
    weather_dict[datetime_idx].append(row.get('dwpt', 0))
    weather_dict[datetime_idx].append(row.get('rhum', 0))
    weather_dict[datetime_idx].append(row.get('prcp', 0))
    weather_dict[datetime_idx].append(row.get('wdir', 0))
    weather_dict[datetime_idx].append(row.get('wspd', 0))
    weather_dict[datetime_idx].append(row.get('pres', 0))
    
    if math.isnan(weather_dict[datetime_idx][4]):
        weather_dict[datetime_idx][4] = 0.0
```

```python
weather_dict
```

```python
parking_size = []
num_of_muni = []
temp = []
dwpt = []
rhum = []
prcp = []
wdir = []
wspd = []
pres = []

for idx, row in lyft_data.iterrows():
    zipcode = row['zipcode']
    datetime_idx = row["date"].replace('-', '')
    time = str(row["time"])
    if len(time) == 1:
        time = '0' + time
    datetime_idx = datetime_idx + '_' + time
        
    # append parking and # of muni data 
    parking_size.append(parking_dict.get(zipcode, 0))
    num_of_muni.append(muni_dict.get(zipcode, 0))
    
    # append weather data
    weather_list = weather_dict.get(datetime_idx, 0)
    temp.append(weather_list[0])
    dwpt.append(weather_list[1])
    rhum.append(weather_list[2])
    prcp.append(weather_list[3])
    wdir.append(weather_list[4])
    wspd.append(weather_list[5])
    pres.append(weather_list[6])
    
lyft_data['parking_size'] = parking_size
lyft_data['num_of_muni'] = num_of_muni
lyft_data['temp'] = temp
lyft_data['dwpt'] = dwpt
lyft_data['rhum'] = rhum
lyft_data['prcp'] = prcp
lyft_data['wdir'] = wdir
lyft_data['wspd'] = wspd
lyft_data['pres'] = pres

lyft_data.head()
```

```python
lyft_data.to_csv("aggregated_dataset_updated.csv")
```
